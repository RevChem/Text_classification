## Еvidently - библиотека

**Evidently** - это библиотека, которая помогает оценивать, тестировать и мониторить производительности ML-моделей.

На текущий момент Evidently проводит анализ по трем основным направлениями:

- Качество моделей
- Дрейф фичей/таргета
- Качество данных

Технически Evidently может работать на трех уровнях:

1. Интерактивном – т.е. вы проводите тесты и строите графики в ноутбуке.
2. Как часть пайплана обработки данных – вы встраиваете методы Evidently в свой DAG.
3. Как BI-инструмент – отправляйте метрики в BI систему вроде Grafana.

## Дрейф данных

В задачах машинного обучения выделяют два основных типа дрейфа: **дрейф концепции** и **дрейф данных**.

>Здесь и далее:

>$P(Х)$ - безусловная вероятность входных признаков X,

>$P(y|Х)$ — условная вероятность таргетов y при входных признаках Х.

**Дрейф концепции** — это изменение P(y|X), т.е. изменение зависимости целевой переменной y от признаков X. При дрейфе концепции, в отличие от дрейфа данных, распределения признаков могут остаться прежними. Иными словами, дрейф концепции происходит, когда зависимости, которым научилась модель, больше не соблюдаются. Это делает модель менее точной или даже устаревшей. Для детекции дрейфа концепции нужно обнаружить изменение в распределении целевой переменной y или изменение зависимостей y от X.

**Дрейф данных** — это изменение P(X). По сути, это изменение входных данных, а не зависимостей между признаками X и целевой меткой y. Однако изменение P(X) может сопровождаться изменением P(y|X). Для детекции дрейфа данных нужно обнаружить изменение в распределении входных признаков X.

**Cкорость дрейфа.** Чаще всего дрейф происходит постепенно и его трудно заметить. Однако внезапные или радикальные внешние изменения могут привести к резкому дрейфу. Яркие примеры таких изменений - это пандемия Covid или внешнеэкономические санкции. Однако к резкому дрейфу могут привести и менее значительные изменения, например, закрытие одного магазина перенаправит поток клиентов в другой, что приведёт к резкому дрейфу признака “посещаемость”.

### Основные подходы к обнаружению дрейфа

Существует множество методов для детекции дрейфа данных и дрейфа концепции. С точки зрения оперативности детекции дрейфа, их можно разделить на 2 типа:

**Методы обнаружения дрейфа в потоковых данных** могут и получать на вход по одному элементу за раз (хотя рациональнее будет подавать данные небольшими пакетами) и сигнализировать о возможном дрейфе практически сразу, как он произошёл. Это достигается за счёт того, что данные методы хранят статистики данных за предыдущие временные срезы и обновляют их с получением нового элемента, принимая решение о наличии дрейфа.

**Методы обнаружения дрейфа в накопленных данных** получают на вход уже собранные наборы данных, поэтому фиксация дрейфа происходит постфактум.

## Что делать с выявленным дрейфом?

Существуют разные подходы к реакции на дрейф данных, но начать следует с **анализа первопричины**.

1. **Откалибровать или перестроить модель**. Удалить или скорректировать дрейфующие признаки и сохранить модель. Один из вариантов работы с дрейфующими признаками, состоит в том, чтобы просто удалить их из рассмотрения моделью или обработать, например, заменить их средним значением. Это не требует сбора дополнительных размеченных данных или переобучения, но может привести к некоторой потере производительности. Однако, если эти потери ниже, чем ожидаемая деградация из-за дрейфа, удаление может быть хорошим вариантом.

- **Удалить или скорректировать дрейфующие признаки**. В некоторых случаях можно удалить дрейфующий признак и надеяться, что переобучение использует альтернативные признаки для достижения аналогичной прогностической способности.

- **Обновить модель.** Переобучить модель на новых данных.

- **Обучить модель на взвешенных данных.** Вместо того, чтобы отбрасывать старые данные, используйте вес, обратно пропорциональный давности данных.

- **Изменить цель предсказания.** Например, переключитесь с еженедельного на ежедневный прогноз или замените регрессионную модель моделью классификации с разбиением по категориям от «высокого» к «низкому».

2. Рассмотрите решение, не связанное с ML. Например, примените поверх модели бизнес-логику, основанную на понимании причинно-следственных связей, сути процесса или экспертных знаний.

3. Ничего не делайте и продолжайте использовать ту же обученную модель. В большинстве случаев это будет наивный подход, однако он может быть рассмотрен в качестве варианта по нескольким причинам:

- изменение качества работы модели может находиться в допустимых границах;
- вы получите время на сбор меток и возможность принять более взвешенное решение;
- обнаруженный дрейф может быть временным.

## Evidently применительно к табличным данным

Evidently на вход принимает два набора данных: эталонный **(reference)** и текущий **(current)**. Эталонный - это тот, с которым мы производим сравнение и на который в идеале должны быть похожи другие датасеты. Обычно это датасет, на котором производилось обучение. Текущий датасет с т.з. машинного обучения это либо тестовый датасет, либо датасет для предсказания.

>Для некоторых тестов достаточно одного набора данных.

Для анализа Evidently предоставляет две основные абстракции: Наборы тестов **(TestSuite)** и Отчеты **(Report)**:

- **Наборы тестов** помогают сравнить два набора данных структурированным образом и в основном предназначены для автоматического применения. Каждый набор содержит несколько отдельных тестов. Каждый тест оценивает конкретную метрику и возвращает результат: прошел/не прошел.

- **Отчеты** же в основном предназначены для интерактивного выполнения в ноутбуках и исследования данных. Вместо простого прохождения тестов они рассчитывают различные показатели и создают дашборды с богатым визуальным оформлением.

### Требования к данным при применении библиотеки Evidently

- **Два набора данных.** Вам всегда нужно передавать два набора данных: текущий (набор данных, оцениваемый на предмет дрейфа) и эталонный (набор данных, который служит эталоном).

- **Непустые столбцы.** Чтобы оценить дрейф данных или прогнозов в наборе данных, вам необходимо убедиться, что столбцы, которые вы проверяете на дрейф, не пустые. Если эти столбцы пусты либо в справочных, либо в текущих данных, Evidently не рассчитает дрейф распределения и выдаст ошибку.

- **Пустые значения.** Если некоторые столбцы содержат пустые или бесконечные значения (+-np.inf), эти значения будут отфильтрованы при расчете дрейфа распределения в соответствующем столбце.

### Вычисление расстояний

#### Для небольших данных **с <= 1000** наблюдений в эталонном наборе данных:

1. Для числовых столбцов **(n_unique > 5):** двухвыборочный тест Колмогорова-Смирнова.

$$
D = \sup_x |F_1(x) - F_2(x)|
$$

- $ F_1(x), F_2(x) $ — эмпирические функции распределения двух выборок.
- $ D $ — максимальное абсолютное отклонение между ними.
- Тест используется для проверки гипотезы о том, принадлежат ли две выборки одному распределению.
- Непараметрический, подходит для непрерывных распределений.

2. Для категориальных столбцов или числовых столбцов с **n_unique <= 5**: тест хи-квадрат.

$$
\chi^2 = \sum_{i=1}^{k} \frac{(O_i - E_i)^2}{E_i}
$$

- $ O_i $ — наблюдаемые частоты,
- $ E_i $ — ожидаемые частоты,
- $ k $ — количество категорий.
- Используется для проверки различий между наблюдаемым и ожидаемым распределением категориальных данных.
- Часто применяется в A/B тестировании и анализе признаков.

3. Для бинарных категориальных признаков **(n_unique <= 2):** тест пропорциональной разницы для независимых выборок на основе Z-оценки.

$$
Z = \frac{\hat{p}_1 - \hat{p}_2}{\sqrt{\hat{p}(1 - \hat{p}) \left( \frac{1}{n_1} + \frac{1}{n_2} \right)}}
$$
где:
$$
\hat{p} = \frac{n_1 \hat{p}_1 + n_2 \hat{p}_2}{n_1 + n_2}
$$

- $ \hat{p}_1, \hat{p}_2 $ — доли успехов в первой и второй выборке,
- $ n_1, n_2 $ — размеры выборок.
- Позволяет сравнивать доли в двух группах (например, CTR, конверсии).
- Предполагает нормальность при больших выборках.

#### Для больших данных **с > 1000** наблюдений в справочном наборе данных:

1.  Для числовых столбцов **(n_unique > 5):** расстояние Вассерштейна .

**Формула (для одномерного случая):**
$$
W(P, Q) = \int_{-\infty}^{\infty} |F_P(x) - F_Q(x)| dx
$$

**Для дискретных распределений:**

$$
W(P, Q) = \sum_{x} |F_P(x) - F_Q(x)|
$$

- $ F_P(x), F_Q(x) $ — функции распределения двух выборок.
- Интуитивно: «работа», необходимая для преобразования одного распределения в другое.
- Чувствителен к порядку значений — хорош для числовых признаков.
- Используется в Evidently AI для обнаружения дрейфа.

2. Для категориальных столбцов или числовых с **(n_unique <= 5)**: расхождение Дженсена—Шеннона .

$$
\text{JSD}(P \| Q) = \frac{1}{2} D_{KL}(P \| M) + \frac{1}{2} D_{KL}(Q \| M), \quad \text{где } M = \frac{1}{2}(P + Q)
$$

- $ D_{KL} $ — относительная энтропия (Kullback–Leibler divergence),
- $ P, Q $ — два распределения,
- $ M $ — среднее распределение.
- Мера симметричная и всегда конечная.
- Используется для сравнения распределений категориальных или числовых признаков.


> Все тесты по умолчанию используют уровень достоверности 0,95. Оценка дрейфа — это P-значение. **(=< 0,05 означает дрейф)**.

> Все метрики используют порог = 0,1 по умолчанию. Оценка дрейфа — это расстояние/расхождение. **(>= 0,1 означает дрейф).**

> Более подробно на https://docs.evidentlyai.com/metrics/customize_data_drift

```python
# Функция, для генерации html-отчетов по дрейфу данных между двумя датасетами
import os
from datetime import datetime
from evidently.report import Report
from evidently.metrics import DatasetDriftMetric, DataDriftTable
from evidently.metric_preset import DataDriftPreset, TargetDriftPreset

def reports(data, new_data, target, directory):

    os.makedirs(directory, exist_ok=True)
    # DataDriftPreset: общая информация о дрейфе признаков
    # TargetDriftPreset: информация о дрейфе целевой переменной
    report = Report()(metrics=[
        DataDriftPreset(),
        TargetDriftPreset()
    ])

    report.run(reference_data = data, current_data = new_data)
    date = datetime.now().strftime('%Y-%m-%d')
    report.save_html(f'{directory}/report_{date}.html')

    # DataDriftTable: таблица с информацией о дрейфе каждого признака
    # DatasetDriftMetric: метрика, указывающая, произошёл ли значительный дрейф
    column_report = Report(metrics=[
        DataDriftTable(),
        DatasetDriftMetric()
    ])
    column_report.run(reference_data = data, current_data = new_data)
    column_report.save_html(f'{directory}/column_{date}.html')
```

На текущий момент доступны следующие отчеты:

- Data Drift - сравнивает распределения входных признаков.
- Data Quality - считает базовые статистики по фичам.
- Numerical Target Drift - обнаруживает изменения в распределении числового таргета.
- Categorical Target Drift - то же самое что и числового, только для категориального таргета.
- Regression Performance - анализирует производительность регрессионной модели и ее ошибки.
- Classification Performance - анализирует производительность и ошибки модели классификации. Работает как для бинарных, так и для мультиклассовых моделей.
- Probabilistic Classification Performance - анализирует производительность модели классификации с т.з. предсказанных вероятностей, проверяет качество калибровки модели и ее ошибки.
